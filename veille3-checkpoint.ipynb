```python
import pandas as pd
pd.set_option('max_rows', 5)
import numpy as np
reviews = pd.read_csv("C:\Python_Practice\Data_Science\covid.csv")
```

# Maps

```python
review_points_mean = reviews.TotalRecovered.mean()
reviews.TotalRecovered.map(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.TotalRecovered - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

```python
review_points_mean = reviews.TotalRecovered.mean()
reviews.TotalRecovered - review_points_mean
```
